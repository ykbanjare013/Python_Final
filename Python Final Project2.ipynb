```python
counts2=df["modelDesc"].value_counts().reset_index().rename(columns={"index":"Model","modelDesc":"Units_sold"})
counts2.head()
```

```python
dict_df = counts2.set_index('Model')['Units_sold'].to_dict()
d1= dict(list(dict_df.items())[0: 10]) 
```

```python
from pywaffle import Waffle
import pandas as pd
import matplotlib.pyplot as plt

data = d1


fig = plt.figure(
    FigureClass=Waffle,
    rows=10,  # Number of rows in the Waffle chart grid
    columns=40,  # Number of columns in the Waffle chart grid
    values=data,  # Use the 'Count' column as values
    #labels=list(data['Category']),  # Use the 'Category' column as labels
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
)

# Add title
plt.title("Waffle Chart from DataFrame")

# Show the chart
plt.show()
```
